```python
df = pd.read_excel("masterdata(xls).xlsx")

df_places = pd.read_excel('ばしょ.xlsx')
places = list(df_places['place'])

place_mask = []
for place in df['Place']:
    judge = place not in places
    place_mask.append(judge)
    
print(place_mask)
```

```python

df_result = df.loc[place_mask, :]
```
